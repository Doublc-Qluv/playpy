### <font color="0000ff">2. 整数序列求和。 用户输入一个正整数N，计算从1到N（包含）相加之后的结果。</font>


```python
N = int(input("请输入整数N:"))
sum_1toN = 0
for i in range(1, N+1):
    sum_1toN += i
print("The sum from 1 to N is :", sum_1toN)
```

### <font color="0000ff">4. 阶乘计算。计算$1!+2!+3!+\dots+10!$的结果。</font> 




```python
factorial_sum, factorial = 0,1
for i in range(1, 11):
    factorial *= i
    factorial_sum += factorial
print("The sum of the factorial of 1 to 10: {}".format(factorial_sum))
```

### <font color="0000ff">5. 猴子吃桃问题。猴子第一天摘下若干个桃子，当即吃了一半，还不过瘾，又多吃了一个；第二天早上又将剩下的桃子吃掉一半，又多吃了一个。以后每天早上都吃了前一天剩下的一半多一个。到第五天早上想再吃时，只剩下一个桃子。编程计算猴子第一天共摘多少桃子。</font>



```python
n = 1
for i in range(4, 0 -1):
    n = (n+1) << 1
print(n)
```

### <font color="0000ff">6. 健康食谱输出。列出 5 中不同的食材，请输出它们可能组成的所有菜式名称。</font>


```python
diet = ['tomato', 'broccoli', 'cucumber', 'steak', 'shrimp']
for x in range(0, 5):
    for y in range(0, 5):
        if not(x == y):
            print("{} {}".format(diet[x], diet[y]))
```

### <font color="0000ff">7. 五角星的绘制。绘制一个红色的五角星图形。</font>


```python
import turtle as t

t.fillcolor('red')
t.begin_fill()
while True:
    t.forward(200)
    t.right(144)
    if abs(t.pos()) < 1:
        break
t.end_fill()
```

### <font color="0000ff">8. 绘制一个太阳花图形。</font>



```python
import turtle as t
t.color('red', 'yellow')
t.begin_fill()
while True:
    t.forward(200)
    t.left(170)
    if abs(t.pos()) < 1:
        break
t.end_fill()
t.done
```

### <font color="0000ff">9. 绘制一个螺旋线图形。</font>



```python
import turtle as t

t.pensize(2)
colors = ['red', 'green', 'blue', 'yellow']
for x in range(100):
    t.pencolor(colors[x % 4])
    t.forward(2*x)
    t.left(90)
```

### <font color="0000ff">10.绘制一个复杂的彩色螺旋线。</font>


```python
import turtle as t
import time
t.pensize(2)
t.bgcolor('black')
colors = ['red', 'green', 'purple', 'blue']
t.tracer(False)
for x in range(400):
    t.forward(2*x)
    t.color(colors[x % 4])
    t.left(91)
t.tracer(True)
time.sleep(20)
```
